```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#read xls sheet
goals=pd.read_excel(open('UEFA Champions League 2016-2022 Data.xlsx', 'rb'),
              sheet_name='goals')  
```

```python
#read xls sheet
players=pd.read_excel(open('UEFA Champions League 2016-2022 Data.xlsx', 'rb'),
              sheet_name='players')  
```

```python
#read xls sheet
matches=pd.read_excel(open('UEFA Champions League 2016-2022 Data.xlsx', 'rb'),
              sheet_name='matches')  
```

```python
#just to check players who have no team, will be droping them

players[players['TEAM'].isna()]
players[players['TEAM'].isnull()]
```

```python
#not relevant
#find player who have no weight and height

players[players['WEIGHT'].isnull() & players['HEIGHT'].isnull() ]
```

```python
#cleaning up,remove playerw with no team
players = players.dropna(subset=['TEAM'])
players
```

```python
#testing  addiing of colums
#value + nan  = nan

a = ['a','c']
b = ['b',]
sa= pd.Series(a)
sb= pd.Series(b)
sa + '_' + sb
```

```python
#need to fill the first name with empty string since we are going combine to get full name
# nan + anythin = nan

players['FIRST_NAME'] = players['FIRST_NAME'].fillna('')
```

```python
#cleaning up - remove rows with no pid
goals = goals.dropna(subset=['PID'])
goals
```

```python
#now combine the two dataframe to become 1 using left join with pid

pg=pd.merge(goals,players, left_on='PID' , right_on ='PLAYER_ID', how="left" )
```

```python
pg
```

```python
# cleaning up
#after merging, there will still be rows where there is still no pid.
#goals sheet contains but player sheet doest not have such player
#let drop them

pg = pg.dropna(subset=['PLAYER_ID'])
pg
```

```python
#to get info from match date, we need to get info from match sheet and joined
#just a inner join will do

pgm=pd.merge(pg,matches, left_on='MATCH_ID' , right_on ='MATCH_ID')



```

```python
# notice 18 DATE_TIME  is string
pgm.info()
```

```python
#time is of string type, convert to datetime,since we can perform sort later
#DATE_TIME is converted to datetime from string

pgm['DATE_TIME']=pd.to_datetime(pgm['DATE_TIME'],format='%d-%b-%y %I.%M.%S.%f %p')
# pgm[pgm['SEASON'] == '2021-2022'] .sort_values(['DATE_TIME'],ascending=False)

```

```python
# notice 18 DATE_TIME  is datetime object now
pgm.info()
```

```python
#get rows from season 2021-2022 and sort it within datetime
#will displayed the latest goal scored from the season
latest =pgm[pgm['SEASON'] == '2021-2022'] .sort_values(['DATE_TIME'],ascending=False)
```

```python
#re-arrage
dis=['SEASON','DATE_TIME','PID','TEAM','JERSEY_NUMBER','FIRST_NAME','LAST_NAME','DURATION','GOAL_DESC','HOME_TEAM','AWAY_TEAM','HOME_TEAM_SCORE','AWAY_TEAM_SCORE']
latest[dis].head(20)
```

```python
#get top scorer using groupby + count with full name displayed
```

```python
#display full 
#get full name
fName = players['FIRST_NAME']+' '+players['LAST_NAME']

# the previous action fill with nan with empty string. strip to remove the leading and trailing white space
fName1=[x.strip() for x in fName if not pd.isna(x)]
```

```python
#just to check the list len
len(fName1)
```

```python
#get series/list of player id
#it should contain same amount of player as the above calculation
pid=players['PLAYER_ID']
pid.info
```

```python
#create dictionary using 2 list.
#map the pid withe fname


pDict = dict(zip(pid, fName1))

```

```python
#check dict mapping
pDict
```

```python
#using math.isnan doesnt work

#another check to ensure the value is not na
#probably not needed anymore
updatedDict = {k:v for k,v in pDict.items() if not pd.isna(v)}
```

```python
#count the goal per player using groupby
goalScored=latest.groupby(['PID']).count()
goalScored['total'] =goalScored['GOAL_ID']
goalScored
```

```python
#check column inserted
goalScored.info()
```

```python
#perform drop on eveyrthing except the last row
goalScored.drop(goalScored.columns[0:-1], axis=1, inplace=True)   
```

```python
#check everything is droped
goalScored.info()
```

```python
#map playerName to row accrodding to index
goalScored['playerName']=goalScored.index.map(updatedDict)
```

```python
#note this still contain index
goalScored
```

```python
#prepapred columns to join
pInfo= players[['PLAYER_ID','NATIONALITY','TEAM']]
pInfo
```

```python
#define function of top scorer

def getTopScorer(x:int):
    temp= goalScored.sort_values(['total'],ascending=False).head(x)
    
    #need to reset as it have index
    temp= temp.reset_index()

    #get team and nationality
    pInfo= players[['PLAYER_ID','NATIONALITY','TEAM']]
    
    #do a inner join using merge to combine info
    res1=pd.merge(pInfo,temp, left_on='PLAYER_ID' , right_on ='PID')
    
    #what to display     
    dis=['playerName','NATIONALITY','TEAM','total']
    
    return res1[dis].sort_values(['total'],ascending=False)
    
       
res =getTopScorer(1)
print(res)
    
```

```python
#top 5 or n scorer
res =getTopScorer(10)

#hide the index
#print(res.to_string(index=False))

#reset the index
res.reset_index(drop=True)
```
